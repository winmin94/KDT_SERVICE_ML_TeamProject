```python
import os
import subprocess

stems = input('stems 선택: 2, 4, 5 >>> ')
path = input('파일이 있는 경로를 정해주세요. >>> ')
os.chdir(path)
file_name = input('음악 파일의 이름을 적어주세요. >>> ')

nsfile_name = file_name.replace(' ', '_')

try:
    os.rename(os.path.join(path, file_name+'.mp3'), os.path.join(path, nsfile_name+'.mp3'))
except FileNotFoundError:
    pass

print('기다려주세요.')
spl = ['spleeter', 'separate', '-p', 'spleeter:' + str(stems) + 'stems', '-o', 'output', nsfile_name+'.mp3']
# ['spleeter', 'separate', '-p', 'spleeter:2stems', '-o', 'output', 'my_song.mp3']

try:
    subprocess.run(spl, check=True)
except subprocess.CalledProcessError as e:
    print("오류가 발생하여 분리 작업을 완료하지 못했습니다.")
    print("오류 메시지:", e)
else:
    print("분리 작업이 완료되었습니다.")
```
